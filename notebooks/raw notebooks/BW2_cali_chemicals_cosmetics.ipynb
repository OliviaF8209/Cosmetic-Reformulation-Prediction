```python
## Checking missing values and Futher Removal of columns

# Checking missing values in each column
missing_values = df.isnull().sum() / len(df) * 100  # Calculate the percentage of missing values per column
print("Percentage of missing values per column:\n", missing_values)

# Filtering out columns with over 95% missing values
columns_to_keep = missing_values[missing_values <= 95].index  # Keep columns with 95% or less missing values
df_cleaned = df[columns_to_keep]

print("\nColumns retained after removing those with over 95% missing values:\n", df_cleaned.columns)

```

```python
df_cleaned['HazardType'] = df_cleaned['ChemicalName'].map(hazard_mapping)

# Count the number of each hazard type per product
hazard_counts = df_cleaned.groupby(['ProductName', 'HazardType']).size().unstack(fill_value=0)

# Display the results
print("Hazard Counts by Product:")
print(hazard_counts)
```

```python
print(df_cleaned[['ChemicalName', 'HazardType']].head(10))
print(df_cleaned['HazardType'].value_counts())
```

```python
# Sum hazardous counts for each product
hazardous_counts_per_product = hazard_counts.sum(axis=1).sort_values(ascending=False)
print("Products with Most Hazardous Chemicals:")
print(hazardous_counts_per_product)
```

```python
import matplotlib.pyplot as plt

# Bar plot for the hazard type distribution
hazard_counts.sum().plot(kind='bar', title='Hazard Type Distribution')
plt.xlabel('Hazard Types')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.show()
```

```python
# Filter for confirmed carcinogens
confirmed_carcinogens = df_cleaned[df_cleaned['HazardType'] == 'Confirmed Carcinogen']
print("Confirmed Carcinogen Products:")
print(confirmed_carcinogens[['ProductName', 'ChemicalName']])
```

```python
# 2. Product Reformulation Count: Counting the number of times a product has been reformulated
if 'ProductName' in df.columns:
    df_cleaned['ReformulationCount'] = df_cleaned.groupby('ProductName')['ChemicalDateRemoved'].transform(
        lambda x: x.notnull().sum()
    )
```

```python
# 3. Average Chemical Age in Product: Calculating the average duration each chemical was in the product
# Ensure date columns are in datetime format
df_cleaned['ChemicalCreatedAt'] = pd.to_datetime(df_cleaned['ChemicalCreatedAt'], errors='coerce')
df_cleaned['ChemicalDateRemoved'] = pd.to_datetime(df_cleaned['ChemicalDateRemoved'], errors='coerce')

# Calculate chemical age (if ChemicalDateRemoved is missing, assume the chemical is still active)
df_cleaned['ChemicalAge'] = (df_cleaned['ChemicalDateRemoved'].fillna(pd.Timestamp.now()) - df_cleaned['ChemicalCreatedAt']).dt.days
df_cleaned['AverageChemicalAge'] = df_cleaned.groupby('ProductName')['ChemicalAge'].transform('mean')
```

```python
# 4. Product Lifespan: Calculating the time from InitialDateReported to DiscontinuedDate
df_cleaned['InitialDateReported'] = pd.to_datetime(df_cleaned['InitialDateReported'], errors='coerce')
df_cleaned['DiscontinuedDate'] = pd.to_datetime(df_cleaned['DiscontinuedDate'], errors='coerce')
df_cleaned['ProductLifespan'] = (df_cleaned['DiscontinuedDate'].fillna(pd.Timestamp.now()) - df_cleaned['InitialDateReported']).dt.days
```

```python
# 5. Chemical Frequency by Product Category: Counting occurrences of each chemical by product category

chemical_counts_by_category = df_cleaned.groupby(['PrimaryCategory', 'ChemicalName']).size().reset_index(name='ChemicalCategoryCount')

# Merge the count information back to the original dataframe
df_cleaned = df_cleaned.merge(chemical_counts_by_category, on=['PrimaryCategory', 'ChemicalName'], how='left')

# Display the first few rows to confirm the addition
print(df_cleaned[['PrimaryCategory', 'ChemicalName', 'ChemicalCategoryCount']].head())
```
