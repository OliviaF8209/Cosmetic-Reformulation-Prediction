##**Column Definitions**##

- **CDPHId**: California Department of Public Health (CDPH) internal identification number for the product. The CDPHId may appear more than once if a product has more than one Color/Scent/Flavor, Category, or Chemical Name/CAS number reported.
  
- **ProductName**: Name of the product as entered by the manufacturer, packer, or distributor. This may contain misspellings or other errors.
  
- **CSFId**: CDPH internal identification number for a color, scent, or flavor.

- **CSF**: Color, scent, or flavor as entered by the manufacturer, packer, or distributor. May contain misspellings or errors.

- **CompanyId**: CDPH internal identification number for the company.

- **CompanyName**: Company name as entered by the manufacturer, packer, or distributor. May contain misspellings or errors.

- **BrandName**: Brand name as entered by the manufacturer, packer, or distributor. May contain misspellings or errors.

- **PrimaryCategoryId**: CDPH internal identification number for the category.

- **PrimaryCategory**: Type of product (e.g., Baby Products, Bath Products, Hair Care Products, etc.).

- **SubCategoryId**: CDPH internal identification number for the subcategory.

- **SubCategory**: Type of product within a primary category.

- **CasId**: CDPH internal identification number for the chemical.

- **CasNumber**: Chemical identification number (CAS registry number). Different forms of a chemical may have different CAS numbers.

- **ChemicalId**: CDPH internal identification number for this chemical's record specific to this product.

- **ChemicalName**: Name of the chemical substance in the cosmetic product.

- **InitialDateReported**: Date the product profile was created by the manufacturer, packer, or distributor.

- **MostRecentDateReported**: Date the product profile was last modified. Defaults to the InitialDateReported value.

- **DiscontinuedDate**: Date the product was discontinued, if applicable.

- **ChemicalCreatedAt**: Date the chemical was first reported to CDPH for this product.

- **ChemicalUpdatedAt**: Date the chemical report was last modified. Defaults to the ChemicalCreatedAt value.

- **ChemicalDateRemoved**: Date the chemical was removed from the product (reformulation date), if applicable.

- **ChemicalCount**: Total number of current chemicals reported for this product, excluding those that have been removed.

```python
chemical_counts = df['ChemicalName'].value_counts()

# Get the top 10 hazardous chemicals
top_10_chemicals = chemical_counts.head(10)

# bar graph
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_chemicals.values, y=top_10_chemicals.index, palette='viridis')

plt.title('Top 10 Hazardous Chemicals and Their Frequencies', fontsize=16)
plt.xlabel('Frequency', fontsize=14)
plt.ylabel('Hazardous Chemicals', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

plt.show()
```

```python
product_counts = df.groupby('PrimaryCategory_Clean')['ProductName_Clean'].nunique().reset_index()
product_counts.columns = ['PrimaryCategory', 'UniqueProductCount']

company_counts = df.groupby('PrimaryCategory_Clean')['CompanyName_Clean'].nunique().reset_index()
company_counts.columns = ['PrimaryCategory', 'UniqueCompanyCount']

# Merge the product and company counts into a single DataFrame for easier visualization
summary_counts = pd.merge(product_counts, company_counts, on='PrimaryCategory')

# Stacked Bar Chart
plt.figure(figsize=(12, 6))
bar_width = 0.35  # Width of the bars
x = range(len(summary_counts))


plt.bar(x, summary_counts['UniqueProductCount'], width=bar_width, label='Unique Products', color='b')
plt.bar([p + bar_width for p in x], summary_counts['UniqueCompanyCount'], width=bar_width, label='Unique Companies', color='orange')

plt.xticks([p + bar_width / 2 for p in x], summary_counts['PrimaryCategory'], rotation=45, ha='right')
plt.title('Breakdown of Hazardous Chemicals by Product Category')
plt.ylabel('Count')
plt.xlabel('Product Category')
plt.legend()
plt.tight_layout()
plt.show()
```
